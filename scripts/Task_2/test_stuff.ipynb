```python
import utils
import pandas as pd
```

```python
data = utils.load_jsonl("../../data/v2/docs_dataset_chunked.jsonl")
data = pd.DataFrame(data)
```

```python
# Get all the chunks with a length of > 120 words
data = data.loc[data["text"].map(lambda text: len(text.split()) >= 120)]
print("Everything Above 120 :", data.shape[0])
```

```python
doc_data = [pd.DataFrame(group[1]) for group in data.groupby("doc_id")]
display(doc_data[50])
```
